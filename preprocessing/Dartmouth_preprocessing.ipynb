```python
from deepface import DeepFace
import pandas as pd
import os
import csv 
```

```python
emotion_image_pairs = dict()
```

```python
folders = ['40Males', '40Females']
emotions = ['Fear', 'Angry', 'Disgust', 'Happy', 'Neutral', 'Pleased', 'Sad', 'Surprise']
for folder in folders:
    for emotion in emotions:
        directory = folder + '/' + emotion
        files = os.listdir(directory)
        index = 0

        while index < len(files):
            filename = files[index]

            if filename[0] == 'S':
                new_dir = directory + '/' + filename
                new_files = os.listdir(new_dir)
                count = 0

                emotion_image_pairs[filename+'_'+emotion] = [emotion]
                while count < len(new_files):
                    new_file = new_files[count]
                    if new_file[0] == 'S':
                        objs = DeepFace.analyze(img_path = os.path.join(new_dir, new_file), 
                            actions = ['emotion'],
                            enforce_detection=False)
                        
                        emotion_image_pairs[filename+"_"+emotion] += [objs[0]['dominant_emotion']]
                        
                    count+= 1
                
            index += 1
```

```python
fields = ['Actual', 'Far_left_1', 'Far_Left_2', 'Far_Right_1', 'Far_Right_2', 'Front_1', 'Front_2', 'Left_1', 'Left_2', "Right_1", 'Right_2', 'Name']
preds_list = []

for key,val in emotion_image_pairs.items():
    val = val[0:11]
    val += [key]
    preds_list += [val]
```

```python
with open('predictions.csv', 'w') as csvfile:  
    # creating a csv writer object  
    csvwriter = csv.writer(csvfile)  
        
    # writing the fields  
    csvwriter.writerow(fields)  
        
    # writing the data rows  
    csvwriter.writerows(preds_list) 
```
